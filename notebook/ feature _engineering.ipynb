```python
import numpy as np
import pandas as pd
df=pd.read_csv("../data/insurance_cleaned.csv",index_col=0)
df
```

```python
df["age_group"]=pd.cut(df["age"],bins=[0,10,19,39,60,120], labels=["Child","Teen","Adults","Middle_Age","Seniors"])
df["bmi_category"]=pd.cut(df["bmi"],bins=[0,18.5,25,30,100], labels=["Underweight","Normal","Overweight","Obese"])
df.head()
```

```python
df.to_csv("../data/insurance_featured.csv",index="False")
```
